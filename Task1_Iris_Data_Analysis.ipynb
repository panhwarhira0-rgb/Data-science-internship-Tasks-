# Task 1: Iris Dataset Analysis

## **Introduction**
This project explores the Iris dataset using Python.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## **Dataset Understanding**
The dataset contains measurements of iris flowers including sepal length, sepal width, petal length and petal width.

```python
df = sns.load_dataset("iris")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.describe()
```

## **Data Visualization**
Scatter plot, histogram and box plot are used to visualize the data.

**Scatter** **plot**

```python
sns.scatterplot(x="sepal_length", y="petal_length", data=df)
plt.show()
```

**Histogram**

```python
df["sepal_length"].hist()
plt.show()
```

**Boxplot**

```python
sns.boxplot(x=df["sepal_length"])
plt.show()
```

## **Conclusion**
The dataset shows different patterns in flower measurements and helps understand relationships between features.
